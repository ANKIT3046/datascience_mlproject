```
df1['DESCRIPTION'] = df1['DESCRIPTION'].str.replace("[^a-zA-Z#]", " ")
```

```
#import nltk liabrary
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
#step 1:Remove punctuation and special characters
df1['DESCRIPTION'] = df1['DESCRIPTION'].str.replace("[^a-zA-Z#]", " ")

#step 2:Remove stopwords
stop_words = stopwords.words('english')
def remove_stopwords(DESCRIPTION):
    desc_new = " ".join([i for i in DESCRIPTION if i not in 
    stop_words])
    return desc_new
#step 3:remove stopwords from the descriptions
DESCRIPTION = [remove_stopwords(str(b).split()) for b in df1['DESCRIPTION']]
#lower cases
DESCRIPTION = [r.lower() for r in DESCRIPTION]

```

```
import spacy
lem = spacy.load('en', disable=['parser', 'ner'])
def lemmatization(texts, tags=['NOUN', 'ADJ']): 
        results= []
        for sent in texts:
            doc = lem(" ".join(sent)) 
            results.append([token.lemma_ for token in doc if token.pos_ in tags])
        return results
#tokenize 
tokenized_descriptions = pd.Series(DESCRIPTION).apply(lambda x: x.split())
descriptions2 = pd.Series(lemmatization(tokenized_descriptions))
descriptions3 = []
for k in range(len(descriptions2)):
    descriptions3.append(' '.join(descriptions2[k]))
```

```
#Word cloud
from wordcloud import WordCloud 
import matplotlib.pyplot as plt

max_words=100
maximum_font=50
def w_cloud(data,bgcolor):
    
    w_cloud1 = WordCloud(background_color = bgcolor, max_words =    
    max_words,  max_font_size = maximum_font).generate(' '.join(data))
    #wc=WordCloud.generate(' '.join(data))
    plt.figure(figsize = (10, 10), facecolor = None) 
    plt.imshow(w_cloud1) 
    plt.axis("off") 
    plt.tight_layout(pad = 0) 
  
    plt.show() 
    
w_cloud(descriptions3,'white')
```
