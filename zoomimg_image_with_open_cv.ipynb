```
import cv2
from google.colab.patches import cv2_imshow
```

```
image=cv2.imread('images1.jpg')
cv2_imshow(image)
```

```
scalex=0.9
scaley=0.9
scaleup=cv2.resize(image,None,fx=scalex*3,fy=scaley*3,interpolation=cv2.INTER_LINEAR)
```

```
cv2_imshow(scaleup)
```

```
cv2_imshow(image)
```

```
scaleup1=cv2.resize(image[10:175,60:170],None,fx=scalex*5,fy=scaley*5,interpolation=cv2.INTER_LINEAR)
```

```
cv2_imshow(scaleup1)
```
