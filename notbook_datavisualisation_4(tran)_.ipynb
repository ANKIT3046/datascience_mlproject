```
import numpy as np
import matplotlib.pyplot as plt
import requests
import urllib.request
import time
from collections import Counter
```

```
from bs4 import BeautifulSoup
```

```
enter_the_url=str(input())
url=enter_the_url
response=requests.get(url).text
```

```
soup=BeautifulSoup(response,'lxml')
li=[]
for para in soup.find_all('p'):
  li.append(para.text.split())
print(li)
  
```

```
total=0
for i in range(len(li)):
  total+=len(li[i])
print(total)
```

```
lis=[]
for i in li:
  for j in i:
    lis.append(j)
print(lis)
len(lis)
```

```
counts=Counter(lis)
print(counts)
```

```
top_20_word=counts.most_common(20)
print(top_20_word)
```

```
val=list()
val_count=list()
for i in range(len(top_20_word)):
  val.append(top_20_word[i][0])
  val_count.append(top_20_word[i][1])
print(val)
print(val_count)
```

```
plt.pie(val_count,labels=val,autopct='%1.1f%%',radius=2.0)
```

```
plt.bar(val,val_count,width=0.1)
```

```
# count the word whose occurence is more than three times
# create two new list for the occurence of word and its count
l_o=[]
l_c=[]
for i in lis:
  if lis.count(i)>3:
    l_o.append(i)
    l_c.append(lis.count(i))
print(l_o)
print(l_c)
```

```
plt.xlabel('words_appere_more_then 3 times')
plt.ylabel('count_of_occurence')
plt.scatter(l_o,l_c,s=50)
```
